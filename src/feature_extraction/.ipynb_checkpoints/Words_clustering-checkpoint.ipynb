```python
import glob, os
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
from itertools import combinations
from sklearn.manifold import MDS
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
import optuna
```

```python
src_pattern = r'/workspace/Krrish/Silent_Speech/dataset_sony/Normalized_dataset1/recordings/*/*.csv'
features = ["theta", "x", "y"]

print("1. Loading & Normalizing PER SUBJECT...")

files = glob.glob(src_pattern)

X, words, subjects = [], [], []
T = 150

for fp in files:
    try:
        df = pd.read_csv(fp)
        data = df[features].values
        #normalization
        scaler = StandardScaler()
        data = scaler.fit_transform(data)
        
        n_chunks = len(data) // T
        if n_chunks == 0:
            continue

        data = data[:n_chunks * T].reshape(n_chunks, T, len(features))
        X.append(data)

        word = os.path.basename(os.path.dirname(fp))
        subj = os.path.basename(fp).replace(".csv", "")

        words.extend([word] * n_chunks)
        subjects.extend([subj] * n_chunks)

    except Exception as e:
        print("Skipping:", fp, e)

X = np.vstack(X)  # (N, T, 5)

print("Total sequences:", X.shape[0])
```

```python
#label coding 
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(words)
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.long)

dataset = torch.utils.data.TensorDataset(X_tensor, y_tensor)
train_loader = DataLoader(
    dataset,
    batch_size=64, 
    shuffle=True,
    drop_last=True
)

full_loader = DataLoader(
    dataset,
    batch_size=64,
    shuffle=False
)
print("2. Dataset & loaders ready.")
```

```python
class KinematicEncoder(nn.Module):
    def __init__(self, in_ch=4, emb_dim=32):
        super().__init__()

        self.net = nn.Sequential(
            nn.Conv1d(in_ch, 64, kernel_size=7, padding=3),
            nn.BatchNorm1d(64),
            nn.ReLU(),

            nn.Conv1d(64, 128, kernel_size=5, padding=2),
            nn.BatchNorm1d(128),
            nn.ReLU(),

            nn.Conv1d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm1d(256),
            nn.ReLU(),

            nn.AdaptiveAvgPool1d(1)  # removes timing/context
        )

        self.proj = nn.Linear(256, emb_dim)

    def forward(self, x):
        # x: (B, T, C)
        x = x.permute(0, 2, 1)  # (B, C, T)
        h = self.net(x).squeeze(-1)
        z = self.proj(h)
        z = F.normalize(z, dim=1)  # critical for metric geometry
        return z
```

```python
class SupConLoss(nn.Module):
    def __init__(self, temperature=0.05):
        super().__init__()
        self.temperature = temperature

    def forward(self, features, labels):
        """
        features: (B, D), normalized
        labels:   (B,)
        """
        device = features.device
        B = features.shape[0]

        sim = torch.matmul(features, features.T) / self.temperature
        sim = sim - sim.max(dim=1, keepdim=True)[0]  # numerical stability

        labels = labels.view(-1, 1)
        mask = torch.eq(labels, labels.T).float().to(device)
        logits_mask = torch.ones_like(mask) - torch.eye(B, device=device)
        mask = mask * logits_mask

        exp_sim = torch.exp(sim) * logits_mask
        log_prob = sim - torch.log(exp_sim.sum(dim=1, keepdim=True) + 1e-8)

        mean_log_prob_pos = (mask * log_prob).sum(dim=1) / (mask.sum(dim=1) + 1e-8)
        loss = -mean_log_prob_pos.mean()

        return loss
```

```python
def temporal_jitter(x, max_shift=20):
    shift = np.random.randint(-max_shift, max_shift)
    return np.roll(x, shift, axis=0)

def channel_dropout(x, p=0.2):
    mask = np.random.rand(x.shape[1]) > p
    return x * mask
```

```python
model = KinematicEncoder(in_ch=3, emb_dim=48).to(device)
criterion = SupConLoss(temperature=0.06)
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=1e-3,
    weight_decay=1e-4
)

num_epochs = 500

for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0

    for x_batch, y_batch in train_loader:
        #batch to numpy for augmentation
        x_np = x_batch.numpy()  # shape: (B, T, C)

        for i in range(len(x_np)):
            x_np[i] = temporal_jitter(x_np[i], max_shift=20)
            x_np[i] = channel_dropout(x_np[i], p=0.2)

        x = torch.tensor(x_np, dtype=torch.float32).to(device)
        y = y_batch.to(device)

        #criterion loss function
        z = model(x)
        loss = criterion(z, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1} | Loss: {total_loss / len(train_loader):.4f}")
```

```python
model.eval()
features, labels = [], []

with torch.no_grad():
    for x_batch, y_batch in full_loader:
        z = model(x_batch.to(device))
        features.append(z.cpu().numpy())
        labels.append(y_batch.numpy())

features = np.vstack(features)
labels = np.concatenate(labels)
```

```python
centroids, names = [], []

for lbl in np.unique(labels):
    centroids.append(features[labels == lbl].mean(axis=0))
    names.append(label_encoder.inverse_transform([lbl])[0])

centroids = np.array(centroids)
sim = cosine_similarity(centroids)

df = pd.DataFrame(sim, index=names, columns=names)

sns.clustermap(
    df,
    method="ward",
    metric="euclidean",
    cmap="mako",
    figsize=(14, 14),
    dendrogram_ratio=0.15
)

plt.title("Articulatory Phonetic Structure (Metric Learning)")
plt.show()
```

```python
Z = linkage(centroids, method="ward")

plt.figure(figsize=(12, 6))
dendrogram(Z, labels=names, leaf_rotation=90)
plt.title("Hierarchical Phonetic Tree (Jaw Kinematics)")
plt.ylabel("Articulatory Distance")
plt.tight_layout()
plt.show()
```

```python
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=40)
dist = 1 - cosine_similarity(centroids)
Z_2d = mds.fit_transform(dist)

plt.figure(figsize=(8, 8))
for i, w in enumerate(names):
    plt.scatter(Z_2d[i, 0], Z_2d[i, 1])
    plt.text(Z_2d[i, 0]+0.01, Z_2d[i, 1]+0.01, w)

plt.title("Word Geometry (MDS  Distance Preserving)")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
intra, inter = [], []

for i in range(len(centroids)):
    for j in range(len(centroids)):
        d = np.linalg.norm(centroids[i] - centroids[j])
        if i == j:
            continue
        if names[i] == names[j]:
            intra.append(d)
        else:
            inter.append(d)

plt.figure(figsize=(6,4))
plt.hist(intra, bins=30, alpha=0.7, label="Intra-word")
plt.hist(inter, bins=30, alpha=0.7, label="Inter-word")
plt.legend()
plt.title("Distance Distribution (Clustering Validity)")
plt.xlabel("Latent Distance")
plt.ylabel("Count")
plt.show()
```

```python
pca = PCA(n_components=2)
Zp = pca.fit_transform(features)

plt.figure(figsize=(7,7))
for s in np.unique(subjects):
    idx = np.array(subjects) == s
    plt.scatter(Zp[idx,0], Zp[idx,1], alpha=0.2)

plt.title("Speaker Spread (Should Overlap)")
plt.axis("equal")
plt.show()
```

```python
model.eval()

all_embeddings = []
all_labels = []
all_words = []
all_subjects = []

with torch.no_grad():
    for xb, yb in full_loader:
        z = model(xb.to(device)).cpu().numpy()
        all_embeddings.append(z)
        all_labels.append(yb.numpy())

all_embeddings = np.vstack(all_embeddings)   # (N, emb_dim)
all_labels = np.concatenate(all_labels)
all_words = label_encoder.inverse_transform(all_labels)
```
