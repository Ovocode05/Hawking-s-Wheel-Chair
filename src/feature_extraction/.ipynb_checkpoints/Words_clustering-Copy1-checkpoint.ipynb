```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import glob
import os
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Configuration
device = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 128
TEMP = 0.06 # Temperature for Contrastive Loss
EMBED_DIM = 64

# ==========================================
# 1. Dataset Class (with Normalization)
# ==========================================
class KinematicDataset(Dataset):
    def __init__(self, src_pattern):
        self.files = glob.glob(src_pattern)
        self.data = []
        self.labels = []
        self.chunk_len = 150
        
        print(f"Loading files from {src_pattern}...")
        
        for fp in self.files:
            try:
                # Extract Label from Folder Name
                word = os.path.basename(os.path.dirname(fp))
                
                # Load Data
                df = pd.read_csv(fp)
                # Assuming columns are ['theta', 'x', 'y'] based on your notebook
                raw_data = df[['theta', 'x', 'y']].values 
                
                # Normalize PER FILE (Important for robust features)
                scaler = StandardScaler()
                norm_data = scaler.fit_transform(raw_data)
                
                # Chunking
                n_chunks = len(norm_data) // self.chunk_len
                for i in range(n_chunks):
                    chunk = norm_data[i*self.chunk_len : (i+1)*self.chunk_len]
                    
                    # Transpose for PyTorch Conv1d: (Length, Channels) -> (Channels, Length)
                    chunk = chunk.T 
                    
                    self.data.append(chunk)
                    self.labels.append(word)
                    
            except Exception as e:
                continue
                
        # Encode Labels
        self.le = LabelEncoder()
        self.labels_enc = self.le.fit_transform(self.labels)
        self.data = np.array(self.data, dtype=np.float32)
        
        print(f"Loaded {len(self.data)} samples. Classes: {len(self.le.classes_)}")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels_enc[idx]

# ==========================================
# 2. The Model: Kinematic Encoder (CNN)
# ==========================================
class KinematicEncoder(nn.Module): 
    def __init__(self, in_ch=3, emb_dim=64):
        super().__init__()

        self.net = nn.Sequential(
            # Block 1
            nn.Conv1d(in_ch, 64, kernel_size=7, padding=3),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            
            # Block 2
            nn.Conv1d(64, 128, kernel_size=5, padding=2),
            nn.BatchNorm1d(128),
            nn.ReLU(),

            # Block 3
            nn.Conv1d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            
            # Global Pooling
            nn.AdaptiveAvgPool1d(1)
        )
        
        # Projection Head (Mapping to Embedding Space)
        self.projection = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, emb_dim)
        )

    def forward(self, x):
        x = self.net(x)
        x = x.squeeze(-1)
        x = self.projection(x)
        # Normalize vectors to unit sphere (Critical for Contrastive Loss)
        x = F.normalize(x, dim=1)
        return x
```

```python
# ==========================================
# 3. Supervised Contrastive Loss
# ==========================================
class SupConLoss(nn.Module):
    def __init__(self, temperature=0.06):
        super().__init__()
        self.temperature = temperature

    def forward(self, features, labels):
        """
        features: (Batch, Dim) - Normalized
        labels: (Batch)
        """
        device = features.device
        batch_size = features.shape[0]

        # Compute Similarity Matrix
        sim_matrix = torch.matmul(features, features.T) / self.temperature
        
        # Numerical Stability: Shift max value to 0 to prevent exp() overflow
        sim_matrix_max, _ = torch.max(sim_matrix, dim=1, keepdim=True)
        sim_matrix = sim_matrix - sim_matrix_max.detach()

        # Mask for same-class positives
        labels = labels.view(-1, 1)
        mask = torch.eq(labels, labels.T).float().to(device)
        
        # Remove self-contrast (diagonal)
        logits_mask = torch.scatter(
            torch.ones_like(mask), 
            1, 
            torch.arange(batch_size).view(-1, 1).to(device), 
            0
        )
        mask = mask * logits_mask

        # Compute Log Probabilities
        exp_sim = torch.exp(sim_matrix) * logits_mask
        
        # Prevent log(0)
        log_prob = sim_matrix - torch.log(exp_sim.sum(1, keepdim=True) + 1e-6)
        
        # --- THE FIX IS HERE ---
        # Compute the denominator (number of positives)
        denominator = mask.sum(1)
        
        # Add epsilon to denominator to prevent 0/0 -> NaN
        # If denominator is 0, the numerator is also 0 (because mask is 0), so result becomes 0.
        mean_log_prob_pos = (mask * log_prob).sum(1) / (denominator + 1e-6)
        
        loss = -mean_log_prob_pos
        
        # Only average over samples that actually had positives (non-zero loss)
        # This prevents the loss from being diluted by "lonely" samples
        loss = loss.mean()
        
        return loss
# ==========================================
# 4. Training Routine
# ==========================================
def train_encoder():
    # Path
    src_pattern = r'/workspace/Krrish/Silent_Speech/dataset_sony/Normalized_dataset/recordings/*/*.csv'
    
    # Data
    dataset = KinematicDataset(src_pattern)
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    
    # Init
    model = KinematicEncoder(in_ch=3, emb_dim=EMBED_DIM).to(device)
    criterion = SupConLoss(temperature=TEMP)
    optimizer = torch.optim.Adam(model.parameters(), lr=4e-4, weight_decay=1e-4)
    
    print("Starting Training...")
    
    for epoch in range(100): # 100 Epochs usually sufficient
        model.train()
        total_loss = 0
        
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            
            optimizer.zero_grad()
            embeddings = model(x)
            loss = criterion(embeddings, y)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            
        if epoch % 10 == 0:
            avg_loss = total_loss / len(loader)
            print(f"Epoch {epoch} | Loss: {avg_loss:.4f}")
            
    # Save the trained model
    torch.save(model.state_dict(), "kinematic_encoder_weights.pth")
    print("Model Saved.")
    return model, dataset

# Run Training
trained_model, full_dataset = train_encoder()
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def visualize_clusters(model, dataset):
    model.eval()
    loader = DataLoader(dataset, batch_size=256, shuffle=False)
    
    all_embeds = []
    all_labels = []
    
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            z = model(x).cpu().numpy()
            all_embeds.append(z)
            all_labels.append(y.numpy())
            
    all_embeds = np.vstack(all_embeds)
    all_labels = np.concatenate(all_labels)
    
    print("Running t-SNE...")
    tsne = TSNE(n_components=2, perplexity=30, random_state=42)
    z_tsne = tsne.fit_transform(all_embeds)
    
    # Plot
    plt.figure(figsize=(12, 10))
    # Decode labels back to strings for legend
    label_names = dataset.le.inverse_transform(all_labels)
    unique_labels = np.unique(label_names)
    
    for lbl in unique_labels:
        mask = label_names == lbl
        plt.scatter(z_tsne[mask, 0], z_tsne[mask, 1], label=lbl, alpha=0.6, s=15)
        
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
    plt.title("Word Clusters (Kinematic Encoder Output)")
    plt.tight_layout()
    plt.show()

# Run Visualization
visualize_clusters(trained_model, full_dataset)
```

```python
# ==========================================
# 2. Generate PINN Inputs (The Bridge)
# ==========================================
import pandas as pd
import glob
import os
from sklearn.preprocessing import StandardScaler

print("Generating unique embeddings for all instances...")

# Paths
src_pattern = r'/workspace/Krrish/Silent_Speech/dataset_sony/Normalized_dataset/recordings/*/*.csv'
files = glob.glob(src_pattern)

# Storage for PINN
theta_dict = {}       # Target: Raw Theta (Physics Ground Truth)
embed_time_dict = {}  # Input: Time + CNN Embedding (The Driver)

t = torch.linspace(0.0, 1.0, 150, device=device).unsqueeze(-1) # Normalized Time Vector
features = ["theta", "x", "y"] # Input features for CNN

count = 0

for fp in files:
    try:
        # 1. Parse ID (Word::Speaker::File)
        word = os.path.basename(os.path.dirname(fp))
        filename = os.path.basename(fp)
        clean_name = filename.replace('.csv', '')
        speaker = clean_name.split('_')[0]
        
        instance_base_id = f"{word}::{speaker}::{filename}"
        
        # 2. Load & Prepare Data
        df = pd.read_csv(fp)
        
        # A. For PINN Target (Raw Theta)
        theta_raw = df["theta"].values
        
        # B. For CNN Input (Theta, X, Y - Normalized per file)
        cnn_data = df[features].values
        scaler = StandardScaler()
        cnn_data_norm = scaler.fit_transform(cnn_data) 
        
        # 3. Chunking (Length 150)
        n_chunks = len(theta_raw) // 150
        if n_chunks == 0: continue
        
        for i in range(n_chunks):
            chunk_id = f"{instance_base_id}::{i}"
            
            # --- Get Target (Raw Theta) ---
            theta_chunk = theta_raw[i*150 : (i+1)*150]
            theta_dict[chunk_id] = torch.tensor(
                theta_chunk, dtype=torch.float64, device=device
            ).unsqueeze(-1)
            
            # --- Get Input (CNN Embedding) ---
            # 1. Get the chunk
            cnn_chunk = cnn_data_norm[i*150 : (i+1)*150]
            
            # 2. Reshape for PyTorch Conv1d: (Batch=1, Channels=3, Time=150)
            cnn_input = torch.tensor(cnn_chunk, dtype=torch.float32, device=device)
            cnn_input = cnn_input.T.unsqueeze(0) 
            
            # 3. Pass through CNN
            with torch.no_grad():
                embedding = cnn_model(cnn_input) # Output: [1, 64]
                
            # 4. Prepare PINN Input [150, 65]
            # Convert embedding to Double to match PINN precision
            emb_expanded = embedding.repeat(150, 1).double() 
            pinn_input = torch.cat([t, emb_expanded], dim=1)
            
            embed_time_dict[chunk_id] = pinn_input
            count += 1
            
    except Exception as e:
        print(f"Skipping {fp}: {e}")
        continue

print(f"Success! Generated {count} unique embeddings.")
print("The PINN will now see unique vectors for every file, eliminating data leakage.")
```
